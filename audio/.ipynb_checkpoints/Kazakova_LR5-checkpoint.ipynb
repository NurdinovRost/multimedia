```python
# Импорт библиотек
import math
import wave
import struct

# Начинаем писать для каждой ноты отдельную функцию
audio = []
sample_rate = 44100.0

# Пустой звук (тишина 0.5 секунд)
def append_silence(duration_milliseconds=500):
    num_samples = duration_milliseconds * (sample_rate / 1000.0)

    for x in range(int(num_samples)): 
        audio.append(0.0)

    return

# Пустой звук (тишина 1 секунда)
def append_silence_long(duration_milliseconds=1000):
    num_samples = duration_milliseconds * (sample_rate / 1000.0)

    for x in range(int(num_samples)): 
        audio.append(0.0)

    return


# Си второй октавы 0.150 сек
def append_si_sh(
        freq=987.00, 
        duration_milliseconds=150, 
        volume=1.0):

    global audio

    num_samples = duration_milliseconds * (sample_rate / 1000.0)

    for x in range(int(num_samples)):
        audio.append(volume * math.sin(2 * math.pi * freq * ( x / sample_rate )))

    return

# Ре второй октавы 0.5 секунд
def append_re_lo(
        freq=587.20, 
        duration_milliseconds=500, 
        volume=1.0):

    global audio 

    num_samples = duration_milliseconds * (sample_rate / 1000.0)

    for x in range(int(num_samples)):
        audio.append(volume * math.sin(2 * math.pi * freq * ( x / sample_rate )))

    return


# Си второй октавы 0.5 секунд
def append_si_lo(
        freq=987.00, 
        duration_milliseconds=500, 
        volume=1.0):

    global audio 

    num_samples = duration_milliseconds * (sample_rate / 1000.0)

    for x in range(int(num_samples)):
        audio.append(volume * math.sin(2 * math.pi * freq * ( x / sample_rate )))

    return

# Ре второй октавы 0.2 секунды
def append_re_sh(
        freq=587.20, 
        duration_milliseconds=200, 
        volume=1.0):

    global audio 

    num_samples = duration_milliseconds * (sample_rate / 1000.0)

    for x in range(int(num_samples)):
        audio.append(volume * math.sin(2 * math.pi * freq * ( x / sample_rate )))

    return

# Ми второй октавы
def append_mi(
        freq=625.20, 
        duration_milliseconds=200, 
        volume=1.0):

    global audio 

    num_samples = duration_milliseconds * (sample_rate / 1000.0)

    for x in range(int(num_samples)):
        audio.append(volume * math.sin(2 * math.pi * freq * ( x / sample_rate )))

    return

# Фа-диез второй октавы
def append_fa(
        freq=739.20, 
        duration_milliseconds=600, 
        volume=1.0):

    global audio 

    num_samples = duration_milliseconds * (sample_rate / 1000.0)

    for x in range(int(num_samples)):
        audio.append(volume * math.sin(2 * math.pi * freq * ( x / sample_rate )))

    return


# Функция сохранения нашего wav файла
def save_wav(file_name):

    wav_file=wave.open(file_name,"w")

    # Параметры wav
    nchannels = 1

    sampwidth = 2

    nframes = len(audio)
    comptype = "NONE"
    compname = "not compressed"
    wav_file.setparams((nchannels, sampwidth, sample_rate, nframes, comptype, compname))
    
    # wav использует 16 бит, поэтому мы умножаем наш сэмпл на 32767 - максимальное значение для int

    for sample in audio:
        wav_file.writeframes(struct.pack('h', int( sample * 32767.0 )))

    wav_file.close()

    return

# Запускаем функции нот и сохраняем результат
append_si_sh(volume=0.5)
append_re_lo(volume=0.5)
append_silence()
append_re_sh(volume=0.5)
append_si_lo(volume=0.5)
append_silence()
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_fa(volume=0.5)
append_silence_long()

append_si_sh(volume=0.5)
append_re_lo(volume=0.5)
append_silence()
append_re_sh(volume=0.5)
append_si_lo(volume=0.5)
append_silence()
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_fa(volume=0.5)
append_silence_long()

append_si_sh(volume=0.5)
append_re_lo(volume=0.5)
append_silence()
append_re_sh(volume=0.5)
append_si_lo(volume=0.5)
append_silence()
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_fa(volume=0.5)
append_silence_long()

append_si_sh(volume=0.5)
append_re_lo(volume=0.5)
append_silence()
append_re_sh(volume=0.5)
append_si_lo(volume=0.5)
append_silence()
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_fa(volume=0.5)
append_silence_long()

append_si_sh(volume=0.5)
append_re_lo(volume=0.5)
append_silence()
append_re_sh(volume=0.5)
append_si_lo(volume=0.5)
append_silence()
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_re_sh(volume=0.5)
append_mi(volume=0.5)
append_fa(volume=0.5)
append_silence_long()
save_wav("output.wav")
```

```python
# Через библиотеку pydub объединяем полученный файл с другим (неважно каким) + делаем нарастание звука и затихание 
from pydub import AudioSegment

signal = AudioSegment.from_wav("output.wav")

bbb = AudioSegment.from_wav("bbb.wav")

res = signal.overlay(bbb)

ready = res.fade_in(3000).fade_out(3000)

ready.export("ready.wav", format="wav")
```

```python
# Создаем спектрограмму полученной композиции
import numpy
import matplotlib.pyplot as plt
import pylab
from scipy.io import wavfile
from scipy.fftpack import fft


myAudio = "ready.wav"

samplingFreq, mySound = wavfile.read(myAudio)

mySoundDataType = mySound.dtype

mySound = mySound / (2.**15)

mySoundShape = mySound.shape
samplePoints = float(mySound.shape[0])

signalDuration =  mySound.shape[0] / samplingFreq

mySoundOneChannel = mySound[:,0]

timeArray = numpy.arange(0, samplePoints, 1)

#
timeArray = timeArray / samplingFreq

timeArray = timeArray * 1000

plt.plot(timeArray, mySoundOneChannel, color='G')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.show()


mySoundLength = len(mySound)

fftArray = fft(mySoundOneChannel)

numUniquePoints = numpy.ceil((mySoundLength + 1) / 2.0)
fftArray = fftArray[0:numUniquePoints]

fftArray = abs(fftArray)

fftArray = fftArray / float(mySoundLength)

fftArray = fftArray **2

if mySoundLength % 2 > 0: 
    fftArray[1:len(fftArray)] = fftArray[1:len(fftArray)] * 2

else: 
    fftArray[1:len(fftArray) -1] = fftArray[1:len(fftArray) -1] * 2  

freqArray = numpy.arange(0, numUniquePoints, 1.0) * (samplingFreq / mySoundLength);

plt.plot(freqArray/1000, 10 * numpy.log10 (fftArray), color='B')
plt.xlabel('Frequency (Khz)')
plt.ylabel('Power (dB)')
plt.show()
```
