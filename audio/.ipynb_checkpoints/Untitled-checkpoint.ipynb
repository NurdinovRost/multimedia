```python
# Импорт библиотек
import math
import wave
import struct

# Начинаем писать для каждой ноты отдельную функцию
audio = []
sample_rate = 44100.0

freq_array = np.array([261.63, 293.66, 329.63, 349.23, 392.00, 440.00, 493.88])

# Пустой звук (тишина 0.5 секунд)
def append_silence(duration_milliseconds=500):
    num_samples = duration_milliseconds * (sample_rate / 1000.0)

    for x in range(int(num_samples)): 
        audio.append(0.0)

    return

# Пустой звук (тишина 1 секунда)
def append_silence_long(duration_milliseconds=1000):
    num_samples = duration_milliseconds * (sample_rate / 1000.0)

    for x in range(int(num_samples)): 
        audio.append(0.0)

    return


def generic_note(freq, duration_milliseconds=1000, volume=1.0, octave=2):
    global audio 
    freq *= octave
    k = 2
    duration_milliseconds *= k
    num_samples = duration_milliseconds * (sample_rate / 1000.0)

    for x in range(int(num_samples)):
        audio.append(volume * math.sin(2 * math.pi * freq * ( x / sample_rate )))

    return

def generic_silence(duration_milliseconds=250):
    num_samples = duration_milliseconds * (sample_rate / 1000.0)

    for x in range(int(num_samples)): 
        audio.append(0.0)

    return

# Функция сохранения нашего wav файла
def save_wav(file_name):

    wav_file=wave.open(file_name,"w")

    # Параметры wav
    nchannels = 1

    sampwidth = 2

    nframes = len(audio)
    comptype = "NONE"
    compname = "not compressed"
    wav_file.setparams((nchannels, sampwidth, sample_rate, nframes, comptype, compname))
    
    # wav использует 16 бит, поэтому мы умножаем наш сэмпл на 32767 - максимальное значение для int

    for sample in audio:
        wav_file.writeframes(struct.pack('h', int( sample * 32768.0 )))

    wav_file.close()

    return

# Запускаем функции нот и сохраняем результат
generic_note(freq=freq_array[1], duration_milliseconds=300)
generic_note(freq=freq_array[3], duration_milliseconds=250)
generic_note(freq=freq_array[1], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=50)
generic_note(freq=freq_array[1], duration_milliseconds=50)
generic_note(freq=freq_array[4], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=150)
generic_note(freq=freq_array[0], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=300)
generic_note(freq=freq_array[5], duration_milliseconds=250)
generic_note(freq=freq_array[1], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=50)
generic_note(freq=freq_array[1], duration_milliseconds=50)
generic_note(freq=freq_array[6], duration_milliseconds=150)
generic_note(freq=freq_array[5], duration_milliseconds=150)
generic_note(freq=freq_array[3], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=150)
generic_note(freq=freq_array[5], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=150, octave=3)
generic_note(freq=freq_array[1], duration_milliseconds=150)
generic_note(freq=freq_array[0], duration_milliseconds=150)
generic_note(freq=freq_array[0], duration_milliseconds=150)
generic_note(freq=freq_array[0], duration_milliseconds=150)
generic_note(freq=freq_array[5], duration_milliseconds=150, octave=1)
generic_note(freq=freq_array[2], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=600)
generic_note(freq=freq_array[1], duration_milliseconds=300)
generic_silence(duration_milliseconds=250)
generic_silence(duration_milliseconds=250)
generic_note(freq=freq_array[1], duration_milliseconds=300)
generic_note(freq=freq_array[3], duration_milliseconds=250)
generic_note(freq=freq_array[1], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=50)
generic_note(freq=freq_array[1], duration_milliseconds=50)
generic_note(freq=freq_array[4], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=150)
generic_note(freq=freq_array[0], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=300)
generic_note(freq=freq_array[5], duration_milliseconds=250)
generic_note(freq=freq_array[1], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=50)
generic_note(freq=freq_array[1], duration_milliseconds=50)
generic_note(freq=freq_array[6], duration_milliseconds=150)
generic_note(freq=freq_array[5], duration_milliseconds=150)
generic_note(freq=freq_array[3], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=150)
generic_note(freq=freq_array[5], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=150)
generic_note(freq=freq_array[0], duration_milliseconds=150)
generic_note(freq=freq_array[0], duration_milliseconds=150)
generic_note(freq=freq_array[0], duration_milliseconds=150)
generic_note(freq=freq_array[5], duration_milliseconds=150, octave=1)
generic_note(freq=freq_array[2], duration_milliseconds=150)
generic_note(freq=freq_array[1], duration_milliseconds=600)
generic_note(freq=freq_array[1], duration_milliseconds=300)
#generic_note(freq=freq_array[5], duration_milliseconds=1000, octave=1)
#generic_note(freq=freq_array[4], duration_milliseconds=1000, octave=1)
# do - 0, re - 1, mi - 2, fa - 3, sol - 4, la - 5, si - 6
#generic_note(freq=freq_array[2], duration_milliseconds=1000, octave=1)
#generic_note(freq=freq_array[1], duration_milliseconds=1000, octave=1)
#generic_note(freq=freq_array[0], duration_milliseconds=1000, octave=1)
save_wav("output.wav")
    
```
