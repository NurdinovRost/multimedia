```python
# Перевод в Grayscale
gray = image[:,:,0]
gray = np.array(gray, dtype = 'uint8')
io.imshow(gray)
```

```python
# Робастное линейное рястяжение
def brle(img, perc):
    y, x = exposure.histogram(img)
    xmin, xmax = int(np.percentile(x, perc)), int(np.percentile(x, 100 - perc))
    new_img = np.array([np.array([
            np.clip(np.uint8((px - xmin) * (255 / (xmax - xmin))), 0, 255)
            for px in row]) for row in img])
    return new_img
res = brle(image, 5)
io.imshow(res)
```

```python
# Гистограмма изображения после рястяжения
hist,bins = np.histogram(res.flatten(),256,[0,256])

plt.hist(res.flatten(),256,[0,256], color = 'black')
plt.xlim([0,256])
plt.show()
```

```python
resfloat = img_as_float(res)
```

```python
# Карта разности
io.imshow(np.abs(image - res))
io.imsave('karta_image-rob.jpg', np.abs(image - res))
```

```python
# Растяжение по каналам
def exp(img):
    y, x = exposure.histogram(img)
    xmin, xmax = int(np.percentile(x, 0)), int(np.percentile(x, 100))
    new_img = np.array([np.array([
            np.uint8((px - xmin) * (255 / (xmax - xmin))) for px in row]) for row in img])
    return new_img
def po_kanalam(img):
    r = exp(img[:, :, 0])
    g = exp(img[:, :, 1])
    b = exp(img[:, :, 2])
    return np.dstack((r, g, b))

rgbexp = po_kanalam(image)
io.imshow(rgbexp)
```

```python
#2.2 Серый мир
av = np.mean(image)
r = (av / np.mean(image[...,0]))
g = (av / np.mean(image[...,1]))
b = (av / np.mean(image[...,2]))
r1 = r*image[...,0]
r2 = g*image[...,1]
r3 = b*image[...,2]
r1 = np.clip(r1, 0, 255)
r2 = np.clip(r2, 0, 255)
r3 = np.clip(r3, 0, 255)
gw = np.dstack((r1, r2, r3))
gw = np.array(gw, dtype = 'uint8')
io.imshow(gw)
```

```python
#2.3 Добавление шума "Соль и перец"
sp = skimage.util.random_noise(gray, mode='s&p', seed=None, clip=True)
io.imshow(sp)
```

```python
# Карта разности исходника и зашумленного
grayfloat = img_as_float(gray)
io.imshow(np.abs(grayfloat - sp))
io.imsave('karta_gray-sp.jpg', np.abs(grayfloat - sp))
```

```python
# Подавляем шум медианным фильтром размера 3
med = median_filter(sp, 3)
io.imshow(med)
```

```python
#2.4 Свертка
def svertka(image, kernel):
    
    kernel = np.flipud(np.fliplr(kernel))    
    output = np.zeros_like(image)            
    
    image_padded = np.zeros((image.shape[0] + 2, image.shape[1] + 2))   
    image_padded[1:-1, 1:-1] = image
    for x in range(image.shape[1]):     
        for y in range(image.shape[0]):
            
            output[y,x]=(kernel*image_padded[y:y+3,x:x+3]).sum()        
    return output
```

```python
# Усреднение
kernel = np.array([[1/9,1/9,1/9],[1/9,1/9,1/9],[1/9,1/9,1/9]])
averaging = svertka(gray, kernel)
io.imshow(averaging)
```

```python
# Автоусреднение
pilgray = Image.open('gray.jpg')
autoav = pilgray.filter(ImageFilter.SMOOTH) 
autoav.save('autoav.jpg')
```

```python
# Сдвиг на 1
kernel = np.array([[0,0,0],[1,0,0],[0,0,0]])
sdvig1 = svertka(gray, kernel)
io.imshow(sdvig1)
```

```python
# Автосдвиг
pilgray = Image.open('gray.jpg')
autosdvig = ImageChops.offset(pilgray, 1,0)
autosdvig.save('autosdvig.jpg')
```

```python
# Гауссов фильтр
kernel = np.array([[0.09,0.12,0.09],[0.12,0.15,0.12],[0.09,0.12,0.09]])
gauss = svertka(gray, kernel)
io.imshow(gauss)
```

```python
# Автогаусс
pilgray = Image.open('gray.jpg')
autogauss = pilgray.filter(ImageFilter.GaussianBlur) 
autogauss.save('autogauss.jpg')
```

```python
# Повышение резкости
kernel = np.array([[0,-0.04,0],[-0.04,2,-0.04],[0,-0.04,0]])
sharpened = svertka(gray, kernel)
io.imshow(sharpened)
```

```python
# Авторезкость
pilgray = Image.open('gray.jpg')
autosharp = pilgray.filter(ImageFilter.SHARPEN) 
autosharp.save('autosharp.jpg')
```
